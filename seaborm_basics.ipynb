Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

```python
'''
Scatter Plot (sns.scatterplot):
Use: To visualize the relationship between two continuous variables. It helps to identify correlations, patterns, and potential outliers in the data.

Box Plot (sns.boxplot):
Use: To display the distribution of a dataset based on a five-number summary (minimum, Q1, median, Q3, and maximum). It is useful for identifying outliers and comparing distributions between different groups.

Histogram (sns.histplot):
Use: To visualize the distribution of a single continuous variable by dividing the data into bins. It helps to understand the central tendency, variability, and shape of the data distribution.

Heatmap (sns.heatmap):
Use: To display the magnitude of values as colors in a matrix form. It is useful for visualizing correlations, missing data, and displaying data tables with color encoding to represent different values.

Pair Plot (sns.pairplot):
Use: To visualize pairwise relationships in a dataset. It creates a matrix of scatter plots for each pair of variables along with histograms or KDE plots on the diagonal, making it useful for exploring correlations and distributions across multiple variables.
'''
```

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
df = sns.load_dataset("fmri")
df.head()
```

```python
sns.lineplot(x=df["timepoint"],y=df["signal"])
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
